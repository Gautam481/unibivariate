```python
def stdNBgraph(dataset):
    #convert to standard normal deviation using formula
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    
    values=[i for i in dataset]
    z_score=[((j-mean)/std)for j in values]
    sns.displot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
stdNBgraph(dataset['salary'])
```

```python
#10 what is the probability density function of a salary from 700000-900000
```

```python
def pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='green') #distplot gives the chart and we are giving datasetfor datas,kde=true is for kernel distribution (curve)
    pyplot.axvline(startrange,color='Red') #range starting lne
    pyplot.axvline(endrange,color='Red') #range ending lne
    #create a sample data
    sample=dataset
    #calculate parameters
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f,Standard Deviatiion=%.3f' %(sample_mean,sample_std))
    #define the distribution(normal distribution)
    dist=norm(sample_mean,sample_std)

    #so outcomes we can use traditional for loop but here we use one liner for loop

    values=[value for value in range(startrange,endrange)]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
pdf_probability(dataset['salary'],700000,900000)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def cal_vif(X):
   #Calculating VIF
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return (vif)
```

```python
cal_vif(dataset[["ssc_p","hsc_p","degree_p","salary","etest_p"]])
```
